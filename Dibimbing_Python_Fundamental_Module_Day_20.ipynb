# **Pseudocode**
**Pseudocode** is the plain English representation of a computer program or algorithm, which specifies the flow and operation of the program.

It is generally used to represent the **structural flow of a program**, and it is not associated with any specific programming language. The same quality makes it a perfect tool to represent algorithms for various problems.

Some points you need to know about *pseudocode* are:
* It is not a programming language.
* It is just a learning and reasoning tool, which is used by programmers and developers to underline how to write the actual code.
* Pseudocode can not be executed or compiled by any compiler, interpreter, or assembler.
* Unlike programming language code, pseudocode does not follow a strict structure and syntax.

# **Conditional**
A program's control flow is the order in which the program's code executes. The control flow of a Python program is regulated by:
* **conditional**
* loops
* function

````
# simple if
a = 3
if a > 0:
    print('A is a positive number')
````

````
# if with else
a = 3
if a < 0:
    print('A is a negative number')
else:
    print('A is a positive number')
````

````
# if with elif
a = 0
if a > 0:
    print('A is a positive number')
elif a < 0:
    print('A is a negative number')
else:
    print('A is zero')
````

````
# nested if
a = 0
if a > 0:
    if a % 2 == 0:
        print('A is a positive and even integer')
    else:
        print('A is a positive number')
elif a == 0:
    print('A is zero')
else:
    print('A is a negative number')
````

````
# if with logical operator
user = 'James'
access_level = 3
if user == 'admin' or access_level >= 4:
        print('Access granted!')
else:
    print('Access denied!')
````

# **Loop**
A program's control flow is the order in which the program's code executes. The control flow of a Python program is regulated by:
* conditional
* **loops**
* function

There are 2 kind of loops:
* for loop
* while loop

````
# simple for loop
print('this is the example of for\n')
a = 'hello world'
b = 1
for i in a:
  print('Here is the letter number #{} with letter {} in {}:'.format(b,i,a))
  b = b + 1
````

````
# for loop with string
numbers = 'python'
for number in numbers:
  print(number, end=' ')
````

````
# for loop with list
numbers = [0, 1, 2, 3, 4, 5]
for number in numbers:
  print(number)
````

````
# for loop with set
numbers = {0, 1, 2, 3, 4, 5}
for number in numbers:
  print(number)
````

````
# for loop with dictionary v1
complex_dict = {
    'first_name':'Jonas',
    'last_name':'Marten',
    'age':27,
    'country':'Indonesia',
    'is_married':True,
    'skills':['JavaScript', 'React', 'Node', 'MongoDB', 'Python'],
    'address':{
        'city':'Jakarta',
        'zipcode':22104
    }
    }
for dict_key in complex_dict:
  print(dict_key)
````

````
# for loop with dictionary v2
complex_dict = {
    'first_name':'Jonas',
    'last_name':'Marten',
    'age':27,
    'country':'Indonesia',
    'is_married':True,
    'skills':['JavaScript', 'React', 'Node', 'MongoDB', 'Python'],
    'address':{
        'city':'Jakarta',
        'zipcode':22104
    }
    }
for dict_key,value_from_key in complex_dict.items():
  print(dict_key,':',value_from_key)
````

````
# for with break
numbers = (0,1,2,3,4,5)
for number in numbers:
    print(number)
    if number == 3:
        break
````

````
# for with continue
numbers = (0,1,2,3,4,5)
for number in numbers:
  print(number)
  if number == 3:
    continue
  print('Next number should be ', number + 1) 
print('loop\'s end')
````

````
# nested for loop
strings = 'python'
numbers = [0, 1, 2, 3, 4, 5]
for word in strings:
  for number in numbers:
    print(word,'and',number)
````

````
# range as a helper to looping
this_is_range = range(0,10,1)
print('raw range:',this_is_range)
print('range in the form of list:',list(this_is_range))
print('range in the form of set:',set(this_is_range))
````

````
# for using range
for number in range(0,11,1):
  print(number)
else:
  print('The loop stops at', number)
````

````
# simple while loop
print('this is the example of while\n')
a = 1
b = 0
while a < 6:
  print(a)
  a = a+1
````

````
# while loop with else
count = 0
while count < 5:
    print(count)
    count = count + 1
else:
    print('this is else')
````

````
# while with break
count = 0
while count < 5:
    print(count)
    count = count + 1
    if count == 3:
        break
````

````
# while with continue
count = 0
while count < 5:
  if count == 3:
    count = count + 1
    continue
  print(count)
  count = count + 1
````
